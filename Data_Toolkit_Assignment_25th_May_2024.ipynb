```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a time-series dataset
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.normal(loc=20, scale=5, size=len(date_range))
humidity = np.random.normal(loc=50, scale=10, size=len(date_range))

# Create a DataFrame
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature
ax1.plot(df['Date'], df['Temperature'], 'b-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'r-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', label='Histogram')

# Compute the PDF values for a range of x-values
x_values = np.linspace(data.min(), data.max(), 1000)
pdf_values = norm.pdf(x_values, loc=0, scale=1)

# Overlay the PDF as a line plot
plt.plot(x_values, pdf_values, color='red', linewidth=2, label='PDF')

# Customize the plot
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

# Show the plot
plt.show()
```

```python
#17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
#origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.


import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrant for each point
quadrant = np.zeros(100, dtype=int)
quadrant[(x > 0) & (y > 0)] = 1  # Quadrant I
quadrant[(x < 0) & (y > 0)] = 2  # Quadrant II
quadrant[(x < 0) & (y < 0)] = 3  # Quadrant III
quadrant[(x > 0) & (y < 0)] = 4  # Quadrant IV

# Create a DataFrame for plotting
data = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': quadrant})

# Set up the plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', hue='Quadrant', data=data, palette='Set1', legend='full')
```

```python

#np.random.randn(100) generates 100 random values from a standard normal distribution for x and y.
#A list quadrants is created to determine the quadrant for each point based on its position relative to the origin.
#A DataFrame df is created to hold the x, y, and quadrant values.
#sns.scatterplot is used to create the scatter plot, with points colored by their quadrant using the hue parameter.
#plt.figure sets the size of the plot.
#The axes are labeled and the title is set using scatter_plot.set.
#A legend is added with plt.legend.
#plt.show displays the plot.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrants
quadrants = []
for xi, yi in zip(x, y):
    if xi >= 0 and yi >= 0:
        quadrants.append('Q1')
    elif xi < 0 and yi >= 0:
        quadrants.append('Q2')
    elif xi < 0 and yi < 0:
        quadrants.append('Q3')
    elif xi >= 0 and yi < 0:
        quadrants.append('Q4')

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrants})

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='bright', legend='full')

# Add axis labels and title
scatter_plot.set(xlabel='X-axis', ylabel='Y-axis', title='Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

```python
# 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'

#np.linspace(0, 4 * np.pi, 100) generates 100 evenly spaced values between 0 and
#4
#𝜋
#4π.
#np.sin(x) calculates the sine of each value in x.
#figure is used to create a Bokeh figure with a specified title and size.
#p.line adds a line renderer to the figure.
#Grid lines are enabled by setting the grid_line_color for the xgrid and ygrid.
#The axis_label properties of xaxis and yaxis are used to label the axes.
#show(p) displays the plot.

import numpy as np
from bokeh.plotting import figure, show

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave Function', width=800, height=400)

# Add a line renderer
p.line(x, y, line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Label the axes
p.xaxis.axis_label = 'X'
p.yaxis.axis_label = 'Y'

# Show the plot
show(p)

```

```python
# 19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
#values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
#Bar Chart'

import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

# Generate some sample categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Define colors for the bars
colors = factor_cmap('Category', palette=Spectral6, factors=categories)

# Create a Bokeh figure
p = figure(width=500, height=500, x_range=categories, title='Random Categorical Bar Chart',
           toolbar_location=None, tools='hover', tooltips=[('Category', '@Category'), ('Value', '@Value')])

# Add the bars
p.vbar(x='Category', top='Value', source=df, width=0.7, color=colors, legend_field='Category')

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'

# Show the plot
show(p)
```

```python
# 20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

#np.linspace(0, 10, 100) generates 100 evenly spaced values between 0 and 10.
#np.random.randn(100).cumsum() generates 100 random values from a standard normal distribution and takes the cumulative sum to create a line with some variation.
#go.Scatter is used to create the line trace.
#go.Layout is used to set the title and axis labels.
#go.Figure combines the data and layout into a figure, which is then displayed with fig.show().

import plotly.graph_objs as go
import numpy as np

# Generate some random data
x = np.linspace(0, 10, 100)
y = np.random.randn(100).cumsum()

# Create a trace
trace = go.Scatter(
    x=x,
    y=y,
    mode='lines',
    name='Random Data'
)

# Create the layout
layout = go.Layout(
    title='Simple Line Plot',
    xaxis=dict(title='X-axis Label'),
    yaxis=dict(title='Y-axis Label')
)

# Create the figure
fig = go.Figure(data=[trace], layout=layout)

# Show the plot
fig.show()
```
